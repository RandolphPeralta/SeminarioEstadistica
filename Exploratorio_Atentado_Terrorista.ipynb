Ahora realizamos un analisis mediante la informacion que tenemos

**Tendencias temporales**: Analiza la distribución de los incidentes a lo largo del tiempo. ¿Hay algún patrón estacional o tendencia a lo largo de los años, meses o días?


```python
import pandas as pd
import matplotlib.pyplot as plt

#reasignamos los datos del DataFrame CasosAT sin afectarlos
datos1 = CasosAT

# Reordenar las columnas para que coincidan con el formato esperado [año, mes, día]
datos1['Fecha'] = pd.to_datetime(datos1[['Año', 'Mes', 'Día']].rename(columns={'Día': 'day', 'Mes': 'month', 'Año': 'year'}))

# Agrupar por mes y contar la cantidad de incidentes
incidentes_por_mes = datos1.groupby(datos1['Fecha'].dt.to_period('M')).size()

# Graficar la distribución de incidentes por mes
plt.figure(figsize=(10, 6))
incidentes_por_mes.plot(kind='line')
plt.title('Distribución de Incidentes por Mes')
plt.xlabel('Mes')
plt.ylabel('Cantidad de Incidentes')
plt.grid(True)
plt.show()
```

**Distribución geográfica**: Examina la distribución geográfica de los incidentes. ¿Hay áreas o regiones que experimentan más violencia que otras? ¿Cuáles son los municipios más afectados?

```python
import pandas as pd
import matplotlib.pyplot as plt

#reasignamos los datos del DataFrame CasosAT sin afectarlos
datos2 = CasosAT

# Contar la cantidad de incidentes por municipio
incidentes_por_municipio = datos2['Municipio'].value_counts()

# Graficar la distribución de incidentes por municipio
plt.figure(figsize=(12, 8))
incidentes_por_municipio.head(20).plot(kind='bar')
plt.title('Distribución de Incidentes por Municipio (Top 20)')
plt.xlabel('Municipio')
plt.ylabel('Cantidad de Incidentes')
plt.xticks(rotation=45, ha='right')
plt.show()
```

**Modalidades de violencia**: Investiga las diferentes modalidades de violencia reportadas. ¿Cuáles son las más comunes? ¿Hay alguna tendencia en cuanto a las modalidades utilizadas por los presuntos responsables?


```python
import pandas as pd
import matplotlib.pyplot as plt

#reasignamos los datos del DataFrame CasosAT sin afectarlos
datos3 = CasosAT

# Contar la frecuencia de cada modalidad de violencia
modalidades_violencia = datos3['Modalidad'].value_counts()

# Graficar la distribución de modalidades de violencia
plt.figure(figsize=(10, 6))
modalidades_violencia.plot(kind='bar')
plt.title('Modalidades de Violencia Reportadas')
plt.xlabel('Modalidad de Violencia')
plt.ylabel('Frecuencia')
plt.xticks(rotation=45, ha='right')
plt.show()
```

**Presuntos responsables**: Analiza la participación de diferentes grupos o entidades como presuntos responsables de los incidentes. ¿Hay algún grupo que esté más activo que otros? ¿Existen cambios en los presuntos responsables a lo largo del tiempo?

```python
import pandas as pd
import matplotlib.pyplot as plt

#reasignamos los datos del DataFrame CasosAT sin afectarlos
datos4 = CasosAT

# Contar la frecuencia de cada presunto responsable
presuntos_responsables = datos4['Presunto Responsable'].value_counts()

# Graficar la distribución de presuntos responsables
plt.figure(figsize=(10, 6))
presuntos_responsables.plot(kind='bar')
plt.title('Presuntos Responsables de los Incidentes')
plt.xlabel('Presunto Responsable')
plt.ylabel('Frecuencia')
plt.xticks(rotation=45, ha='right')
plt.show()

# Agrupar por año y contar la frecuencia de cada presunto responsable
presuntos_responsables_por_año = datos4.groupby(['Año', 'Presunto Responsable']).size().unstack()

# Graficar la participación de presuntos responsables a lo largo del tiempo
plt.figure(figsize=(12, 8))
presuntos_responsables_por_año.plot(kind='line', marker='o')
plt.title('Participación de Presuntos Responsables a lo largo del Tiempo')
plt.xlabel('Año')
plt.ylabel('Cantidad de Incidentes')
plt.legend(title='Presunto Responsable', loc='upper right')
plt.grid(True)
plt.show()
```

**Impacto y víctimas**: Se va a examinar el número de víctimas y el tipo de impacto reportado en cada incidente. ¿Cuáles son las consecuencias más comunes de estos eventos? ¿Hay diferencias en el impacto dependiendo del tipo de presunto responsable o modalidad de violencia?

Se Graficará la distribución del tipo de impacto reportado en los incidentes y la cantidad total de víctimas asociadas a cada tipo de impacto.

Se Graficará el total de víctimas dividido por el tipo de presunto responsable, diferenciando entre civiles y combatientes.

Se Graficará el total de víctimas dividido por la modalidad de violencia, también diferenciando entre civiles y combatientes.

```python
import pandas as pd
import matplotlib.pyplot as plt

#reasignamos los datos del DataFrame CasosAT sin afectarlos
datos5 = CasosAT

# Obtener el total de víctimas y el tipo de impacto reportado en cada incidente
impacto_victimas = datos5[['Abandono o Despojo Forzado de Tierras', 'Amenaza o Intimidación', 
                          'Ataque Contra Misión Médica', 'Confinamiento o Restricción a la Movilidad', 
                          'Desplazamiento Forzado', 'Extorsión', 'Lesionados Civiles', 'Pillaje', 
                          'Tortura', 'Violencia Basada en Género']].sum()

# Graficar la distribución del tipo de impacto
plt.figure(figsize=(10, 6))
impacto_victimas.sort_values(ascending=False).plot(kind='bar')
plt.title('Impacto y Víctimas Reportadas en los Incidentes')
plt.xlabel('Tipo de Impacto')
plt.ylabel('Cantidad de Víctimas')
plt.xticks(rotation=45, ha='right')
plt.show()

# Calcular el total de víctimas por tipo de presunto responsable
victimas_por_presunto_responsable = datos5.groupby('Presunto Responsable')[['Total Civiles', 'Total Combatientes']].sum()

# Graficar el total de víctimas por tipo de presunto responsable
plt.figure(figsize=(10, 6))
victimas_por_presunto_responsable.plot(kind='bar', stacked=True)
plt.title('Total de Víctimas por Tipo de Presunto Responsable')
plt.xlabel('Presunto Responsable')
plt.ylabel('Total de Víctimas')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Tipo de Víctima', loc='upper right')
plt.show()

# Calcular el total de víctimas por modalidad de violencia
victimas_por_modalidad = datos5.groupby('Modalidad')[['Total Civiles', 'Total Combatientes']].sum()

# Graficar el total de víctimas por modalidad de violencia
plt.figure(figsize=(10, 6))
victimas_por_modalidad.plot(kind='bar', stacked=True)
plt.title('Total de Víctimas por Modalidad de Violencia')
plt.xlabel('Modalidad de Violencia')
plt.ylabel('Total de Víctimas')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Tipo de Víctima', loc='upper right')
plt.show()
```
